```python

import sqlite3

# Conectando ao banco de dados
conexao = sqlite3.connect('lanchonete.db')
cursor = conexao.cursor()

# Inserindo dados na tabela Compra
dados_compra = [
    (1, 1, '2022-01-01', 15.90),
    (2, 2, '2022-01-02', 18.90),
    (3, 3, '2022-01-03', 20.90),
    (4, 4, '2022-01-04', 17.90),
    (5, 5, '2022-01-05', 19.90),
    (6, 6, '2022-01-06', 25.90),
    (7, 7, '2022-01-07', 12.90),
    (8, 8, '2022-01-08', 14.90),
    (9, 9, '2022-01-09', 21.90),
    (10, 10, '2022-01-10', 22.90)
]

# Utilizando uma transação para melhorar a performance
cursor.execute('BEGIN TRANSACTION')

try:
    # Inserindo cada registro de compra
    for item in dados_compra:
        cursor.execute('''
            INSERT INTO Compra (idCliente, idItem, dataCompra, valorTotal)
            VALUES (?, ?, ?, ?)
        ''', item)

    # Efetuando o commit da transação
    conexao.commit()
    print("Registros de compras inseridos com sucesso.")

    # Consulta: Compra por ID com informações de Cliente
    cursor.execute('''
        SELECT Compra.idCompra, Compra.dataCompra, Cliente.nome
        FROM Compra
        INNER JOIN Cliente ON Compra.idCliente = Cliente.idCliente
    ''')
    compra_cliente = cursor.fetchall()
    print("\nConsulta: Compra por ID com informações de Cliente")
    for item in compra_cliente:
        print(item)

    # Consulta: Compra por ID com informações de Cardapio
    cursor.execute('''
        SELECT Compra.idCompra, Compra.valorTotal, Cardapio.nome
        FROM Compra
        LEFT JOIN Cardapio ON Compra.idItem = Cardapio.idItem
    ''')
    compra_cardapio = cursor.fetchall()
    print("\nConsulta: Compra por ID com informações de Cardapio")
    for item in compra_cardapio:
        print(item)

    # Consulta: Compra com ID maior que 5
    cursor.execute('''
        SELECT * FROM Compra WHERE idCompra > 5
    ''')
    compra_id_maior_5 = cursor.fetchall()
    print("\nConsulta: Compra com ID maior que 5")
    for item in compra_id_maior_5:
        print(item)

except Exception as e:
    # Em caso de erro, desfaz a transação
    conexao.rollback()
    print("Ocorreu um erro ao inserir os registros de compras:", str(e))
finally:
    # Fechando a conexão com o banco de dados
    conexao.close()
```
